## Feature engineering and Data Preparation

#### Create boolean flags to identify each type of injury

```python
# Reusable Function (for creating injury flag columns)
def create_injury_flag(df, new_col_name, search_pattern, notes_col='Notes'):
    """
    Adds a boolean flag column to a DataFrame based on a keyword search in a text column.
    """
    print(f"Creating flag column: '{new_col_name}'...")

    if notes_col not in df.columns:
        print(f"  ERROR: Notes column '{notes_col}' not found. Cannot create flag.")
        return df

    # Robustly prepare the notes column before searching
    df[notes_col] = df[notes_col].astype(str).fillna('')

    # Create the boolean flag column
    df[new_col_name] = df[notes_col].str.contains(
        search_pattern,
        case=False,  # Case-insensitive
        na=False,    # Treat NaNs as False
        regex=True   # Use regex matching
    ).astype(int) # Convert final True/False to 1/0 for easier math later

    flag_count = df[new_col_name].sum()
    print(f"  Done. Found {flag_count} instances for '{new_col_name}'.")

    return df

# Define Injury Patterns
# This dictionary holds the new column name and the regex pattern to find it
injury_patterns_to_flag = {
    'is_groin_injury': r'\b(groin|adductor)\b',
    'is_quad_injury': r'\b(quadriceps?|quads?|thigh muscle)\b',
    'is_hamstring_injury': r'\b(hamstring|hammy)\b',
    'is_ankle_injury': r'\b(ankle|ankles)\b',
    'is_back_injury': r'\b(back|spine|lumbar|thoracic|vertebra[e]?)\b',
    'is_concussion_injury': r'\b(concussion|head injur(?:y|ies)|cranial)\b',
    'is_calf_injury': r'\b(calf|gastrocnemius|soleus)\b'
}

# Loop Through the Dictionary and Call the Function
print("\nStarting to create all injury flag columns ")
for column_name, regex_pattern in injury_patterns_to_flag.items():
    injury_data_cleaned = create_injury_flag(
        df=injury_data_cleaned,
        new_col_name=column_name,
        search_pattern=regex_pattern
        # notes_col='Notes' is the default, but you could specify a different one if needed
    )

# Verify the Results
print("\nFinal DataFrame with All Injury Flags")

# Create a list of columns to display for verification
columns_to_verify = [
    'Player',
    'Date',
    'Notes'
] + list(injury_patterns_to_flag.keys())

# Filter for columns that actually exist to prevent KeyErrors
existing_columns_to_verify = [col for col in columns_to_verify if col in injury_data_cleaned.columns]

# Use pandas option context to ensure all columns are displayed
with pd.option_context('display.max_columns', None):
    print(injury_data_cleaned[existing_columns_to_verify].head(10))

# check the value counts for a specific new column
if 'is_ankle_injury' in injury_data_cleaned.columns:
    print("\nValue counts for 'is_ankle_injury':")
    print(injury_data_cleaned['is_ankle_injury'].value_counts())
```

### Create Outcome Variables (for prediction)


*   Recent injury history (1 or 0): groin, quadriceps, hamstring, ankle, back injury, or concussion within 8 weeks of LEM injury (lower extremity muscle strain of the hamstring, quad, calf, groin)
  *   Create is_target_LEMS variable by grouping injury variables for the four lems into one column
  *   Using that variable, whenever value is true subtract back 8 weeks from the date of that injury to look for any injury
      *   If date is within first 8 weeks of season, lookback period is up to start of season as well as the end of previous season (however many weeks in until total period is 8 weeks)
  *   If there is an injury within 8 weeks classify as true in new column titled recent_injury_history
*   Remote injury history (1 or 0): any history of groin, quadriceps, hamstring, ankle, back injury, or concussion before the LEM injury
  *   Using is_target_LEMS variable again, whenever value is true subtrack back all previous injury data from the date of that injury to look for any injury
  *   If there is an injury within all history classify as true in new column titled remote_injury_history
*   Previous total count of lower extremity injuries






####  1. Recent Injury History

```python
injury_data_cleaned['is_target_LEMS'] = (
    injury_data_cleaned['is_hamstring_injury'] |
    injury_data_cleaned['is_quad_injury'] |
    injury_data_cleaned['is_calf_injury'] |
    injury_data_cleaned['is_groin_injury']
)
injury_data_cleaned
```

Standardize player column in injury_data_cleaned to match format of stats_df player column for later merging

```python
import re

def extract_and_standardize_player_name(text_input):
    if pd.isna(text_input):
        return None
    name_str = str(text_input)

    # 1. Remove leading/trailing whitespace
    name_str = name_str.strip()

    # 2. Remove leading "• " if present
    if name_str.startswith("• "):
        name_str = name_str[2:].strip() # Remove the "• " and strip again

    # 3. Handle aliases like "Maximilian Kleber / Maxi Kleber" - take the first part
    if "/" in name_str:
        name_str = name_str.split('/')[0].strip()

    # 4. Attempt to extract name after potential position prefixes (G, F, C)
    # This regex tries to capture a typical name structure.
    # It allows for names like "First Last", "F. Last", "First M. Last", "First Middle Last"
    # It also tries to ignore a leading single character followed by a space (common for positions)
    match = re.match(r"^(?:[A-Z]\s)?([A-Z][a-zA-Z.'-]+(?:\s[A-Z][a-zA-Z.'-]+){0,2}(?:\s(?:Jr\.|Sr\.|III|II|IV))?)", name_str)
    if match:
        extracted_name = match.group(1).strip()
    else:
        # If regex doesn't match, use the string after initial cleaning (steps 1-3)
        # This might happen for less common name structures or if it's already clean.
        extracted_name = name_str

    # 5. Consistent Casing (e.g., Title Case)
    final_name = extracted_name.title()

    # 6. Any other specific replacements or cleaning (e.g., removing specific unwanted characters if any)
    # final_name = final_name.replace('.', '') # Example: if you want to remove periods from "Jr."

    return final_name if final_name else None

if 'Player' in injury_data_cleaned.columns:
    injury_data_cleaned['Player'] = injury_data_cleaned['Player'].apply(extract_and_standardize_player_name)
    # player_identifier_col_in_injury_data = 'Player'
    print("Standardized player names from 'Player' into 'Player'.")
    print(injury_data_cleaned[['Player', 'Player']].head(10)) # Check output
else:
    print("Error: 'Player' column not found for standardization.")
```

```python
lookback_window = pd.Timedelta(weeks=8)

print(f"\nFound {len(df_lems_cases)} target LEMS events. Checking for specific prior precursor injuries...")

player_season_precursors = {}

processed_count = 0
for index, lems_event_row in df_lems_cases.iterrows():
    # standardized player name from injury_data_cleaned
    current_player = lems_event_row['Player']
    lems_date = lems_event_row['Date']

    lems_season = None
    if pd.notna(lems_date):
        if lems_date.month >= 8:
            lems_season = f"{lems_date.year}-{lems_date.year + 1}"
        else:
            lems_season = f"{lems_date.year - 1}-{lems_date.year}"

    # Skip if current_player is None (meaning name extraction failed) or other key info is missing
    if pd.isna(current_player) or not current_player or pd.isna(lems_date) or lems_season is None:
        processed_count += 1
        continue

    lookback_start_date = lems_date - lookback_window

    end_of_2018_19_season = pd.to_datetime("2019-04-10")

    start_of_2019_20_season = pd.to_datetime("2019-10-22")

    end_of_2019_20_season = pd.to_datetime("2020-08-14")

    start_of_2020_21_season = pd.to_datetime("2020-12-22")

    end_of_2020_21_season = pd.to_datetime("2021-05-16")

    start_of_2021_22_season = pd.to_datetime("2021-10-19")

    first_week_lookback_start_date_2 = end_of_2019_20_season - pd.Timedelta(weeks=8)

    first_week_lookback_start_date_3 = end_of_2020_21_season - pd.Timedelta(weeks=8)


    # first week of 2019-20 season (lookback period is end of 2018-19 season date minus 8 weeks)
    if lems_date >= start_of_2019_20_season and lems_date < (start_of_2019_20_season + lookback_window):
      prior_injuries_for_player = injury_data_cleaned[
          (injury_data_cleaned['Player'] == current_player) & # Match on standardized name
          (injury_data_cleaned['Date'] < end_of_2018_19_season)
      ]
    # first week of 2020-21 season (lookback period is end of 2019-21 season date minus 8 weeks)
    elif lems_date >= start_of_2020_21_season and lems_date < (start_of_2020_21_season + lookback_window):
        prior_injuries_for_player = injury_data_cleaned[
            (injury_data_cleaned['Player'] == current_player) & # Match on standardized name
            (injury_data_cleaned['Date'] >= first_week_lookback_start_date_2) &
            (injury_data_cleaned['Date'] < end_of_2019_20_season)
        ]
    # first week of 2021-22 season (lookback period is end of 2020-21 season date minus 8 weeks)
    elif lems_date >= start_of_2021_22_season and lems_date < (start_of_2021_22_season + lookback_window):
          prior_injuries_for_player = injury_data_cleaned[
            (injury_data_cleaned['Player'] == current_player) & # Match on standardized name
            (injury_data_cleaned['Date'] >= first_week_lookback_start_date_3) &
            (injury_data_cleaned['Date'] < end_of_2020_21_season)
          ]
    # all other rows simply look back 8 weeks from that lems occurrence
    else:
          prior_injuries_for_player = injury_data_cleaned[
             (injury_data_cleaned['Player'] == current_player) & # Match on standardized name
             (injury_data_cleaned['Date'] >= lems_date - lookback_window) &
             (injury_data_cleaned['Date'] < lems_date)
          ]


    if not prior_injuries_for_player.empty:
        player_season_key = (current_player, lems_season)
        if player_season_key not in player_season_precursors:
            player_season_precursors[player_season_key] = set()

        # Iterate through list of precursor flag column names
        for precursor_col_name in precursor_flag_columns_in_df:
            if precursor_col_name in prior_injuries_for_player.columns: # Check if column exists
                if prior_injuries_for_player[precursor_col_name].any():
                    player_season_precursors[player_season_key].add(precursor_col_name)
            else:
                print(f"Warning: Column '{precursor_col_name}' not found in prior_injuries_for_player during LEMS processing.")

    processed_count += 1
    if processed_count > 0 and processed_count % 100 == 0:
        print(f"  Processed {processed_count} of {len(df_lems_cases)} LEMS events...")

print("Finished checking specific precursor injuries for LEMS events.")
```

```python
feature_column_mapping = {
    'is_groin_injury': 'recent_groin_injury',
    'is_quad_injury': 'recent_quad_injury',
    'is_hamstring_injury': 'recent_hamstring_injury',
    'is_ankle_injury': 'recent_ankle_injury',
    'is_back_injury': 'recent_back_injury',
    'is_concussion_injury': 'recent_concussion_injury'
}

player_col_in_stats_df = 'Player'

# Initialize the new feature columns
print(f"Initializing new feature columns in stats_df (current shape: {stats_df.shape})...")
for new_feature_col_name_in_stats_df in feature_column_mapping.values():
    stats_df[new_feature_col_name_in_stats_df] = 0 # Reset/Initialize
print("New feature columns initialized.")


if player_col_in_stats_df not in stats_df.columns: # Check the correct player column for stats_df
    raise KeyError(f"Player identifier column '{player_col_in_stats_df}' not found in stats_df.")
if 'Season' not in stats_df.columns:
    raise KeyError(f"'Season' column not found in stats_df.")


print("\nMapping precursor injury information to stats_df...")
processed_rows_count_map = 0
for index, stat_row in stats_df.iterrows():
    player_identifier_from_stats_row = stat_row[player_col_in_stats_df]
    season_from_stats_row = stat_row['Season']

    if pd.isna(player_identifier_from_stats_row) or pd.isna(season_from_stats_row):
        processed_rows_count_map += 1
        continue

    player_season_key_for_lookup = (player_identifier_from_stats_row, season_from_stats_row)

    if player_season_key_for_lookup in player_season_precursors:
        precursors_found_set = player_season_precursors[player_season_key_for_lookup]
        for precursor_flag_key_from_dict, new_feature_col_name in feature_column_mapping.items():
            # precursor_flag_key_from_dict is like 'is_groin_injury_flag'
            if precursor_flag_key_from_dict in precursors_found_set:
                stats_df.loc[index, new_feature_col_name] = 1

    processed_rows_count_map += 1
    if processed_rows_count_map > 0 and processed_rows_count_map % 500 == 0:
        print(f"  Processed {processed_rows_count_map} of {len(stats_df)} rows in stats_df for feature mapping...")

print("Finished mapping precursor injury information.")
```

```python
# This column is True if any of the relevant flags are True for an injury record
existing_relevant_flags = [col for col in lems_precursor_flag_columns_in_df if col in injury_data_cleaned.columns]
if not existing_relevant_flags:
    raise ValueError("None of the lems_precursor_flag_columns_in_df found in injury_data_cleaned. Please check column names.")
if len(existing_relevant_flags) < len(lems_precursor_flag_columns_in_df):
    print(f"Warning: Not all lems_precursor_flag_columns_in_df found. Using: {existing_relevant_flags}")

injury_data_cleaned['is_count_relevant_prior_injury'] = injury_data_cleaned[existing_relevant_flags].any(axis=1)

# Determine Season for LEMS Events in df_lems_cases
def get_season(date):
    if pd.isna(date):
        return None
    if date.month >= 8:
        return f"{date.year}-{date.year + 1}"
    else:
        return f"{date.year - 1}-{date.year}"

df_lems_cases['Season'] = df_lems_cases['Date'].apply(get_season)
# Drop LEMS cases where season could not be determined (if any)
df_lems_cases.dropna(subset=['Player', 'Season', 'Date'], inplace=True)


# Find First LEMS Date per Player-Season
first_lems_dates_df = df_lems_cases.loc[
    df_lems_cases.groupby(['Player', 'Season'])['Date'].idxmin()
][['Player', 'Season', 'Date']].rename(columns={'Date': 'First_LEMS_Date_In_Season'})

# Calculate previous_injury_count
player_season_counts = []

for _, row in first_lems_dates_df.iterrows():
    player = row['Player']
    season = row['Season']
    first_lems_date = row['First_LEMS_Date_In_Season']

    # Filter injury_data_cleaned for this player's injuries BEFORE their first LEMS in this season
    prior_injuries = injury_data_cleaned[
        (injury_data_cleaned['Player'] == player) &
        (injury_data_cleaned['Date'] < first_lems_date) &
        (injury_data_cleaned['is_count_relevant_prior_injury'] == True) # Count only relevant types
    ]

    count = len(prior_injuries)
    player_season_counts.append({
        'Player': player,
        'Season': season,
        'previous_injury_count': count
    })

prior_counts_df = pd.DataFrame(player_season_counts)

# Merge into stats_df
if not prior_counts_df.empty:
    # Assuming stats_df has 'Player' and 'Season' columns matching the format
    stats_df = pd.merge(stats_df, prior_counts_df, on=['Player', 'Season'], how='left')
    # Fill NaN with 0 for players/seasons in stats_df that didn't have a LEMS
    # or had no prior relevant injuries before their first LEMS.
    stats_df['previous_injury_count'] = stats_df['previous_injury_count'].fillna(0).astype(int)
else:
    # If no LEMS events or no prior injuries, add the column with zeros
    stats_df['previous_injury_count'] = 0
    print("No prior injury counts were calculated (prior_counts_df is empty). 'previous_injury_count' column added as 0 to stats_df.")


print("Finished calculating and merging 'previous_injury_count' into stats_df.")
# print(stats_df.head())
# print(stats_df[stats_df['previous_injury_count'] > 0].head())
```

## K-Means Clustering

### Clustering with all injury variables

Goal: Among the player-seasons that did result in a LEMS, are there different types or profiles? Are there distinct data-driven 'pathways' that lead to a LEMS?

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# Select injury columns as features for clustering
feature_columns_for_clustering = [
    'recent_groin_injury',
    'recent_quad_injury',
    'recent_hamstring_injury',
    'recent_ankle_injury',
    'recent_back_injury',
    'recent_concussion_injury',
    'remote_groin_injury',
    'remote_quad_injury',
    'remote_hamstring_injury',
    'remote_ankle_injury',
    'remote_back_injury',
    'remote_concussion_injury',
    'previous_injury_count'
]

X_start = lems_seasons_df[feature_columns_for_clustering].copy()

# Standardize X
scaler = StandardScaler()

X = scaler.fit_transform(X_start)
```

```python
# Determine number of clusters for k-means

# Elbow Method
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, n_init='auto', random_state=42)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)

plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

How to Interpret: Find the point where adding another cluster doesn't give you a significantly better fit. Looking for the "elbow" in the curve – the point where the rate of decrease sharply changes.




```python
# MODIFIED FROM: https://scikit-learn.org/stable/auto_examples/cluster/plot_kmeans_silhouette_analysis.html

from sklearn.metrics import silhouette_score, silhouette_samples
import matplotlib.cm as cm
import numpy as np

fig, axes = plt.subplots(3, 3, figsize=(15, 12))  # 3x3 grid
axes = axes.flatten()  # Flatten to 1D array for easy indexing

for idx, n_clusters in enumerate(range(2, 11)):
    ax = axes[idx]  # Select subplot

    # Fit KMeans and predict labels
    kmeans = KMeans(n_clusters=n_clusters, random_state=42, n_init='auto')
    labels = kmeans.fit_predict(X)

    # Compute silhouette scores
    silhouette_avg = silhouette_score(X, labels)
    sample_silhouette_values = silhouette_samples(X, labels)

    y_lower = 10
    for i in range(n_clusters):
        # Aggregate silhouette scores for the current cluster
        ith_cluster_silhouette_values = sample_silhouette_values[labels == i]
        ith_cluster_silhouette_values.sort()

        size_cluster_i = ith_cluster_silhouette_values.shape[0]
        y_upper = y_lower + size_cluster_i

        color = cm.nipy_spectral(float(i) / n_clusters)
        ax.fill_betweenx(np.arange(y_lower, y_upper),
                         0, ith_cluster_silhouette_values,
                         facecolor=color, edgecolor=color, alpha=0.7)

        y_lower = y_upper + 10  # Add space between clusters

    ax.set_title(f"{n_clusters} Clusters", fontsize=10)
    ax.set_xlabel("Silhouette coefficient", fontsize=9)
    ax.set_ylabel("Cluster", fontsize=9)

    # Draw a vertical line for the average silhouette score
    ax.axvline(x=silhouette_avg, color="red", linestyle="--")

    # Annotate the silhouette score on the plot
    ax.text(0.6, 0.9, f"Avg: {silhouette_avg:.3f}", transform=ax.transAxes, color="red", fontsize=9)

    ax.set_yticks([])  # Remove y-axis labels
    ax.set_xticks(np.arange(-0.1, 1.1, 0.2))

# Adjust layout
fig.suptitle("Silhouette Plots for Different Cluster Numbers", fontsize=14)
fig.tight_layout(rect=[0, 0, 1, 0.98])  # Adjust to fit suptitle

plt.show()
```

### Significance Visual (Pairwise p-value Heatmap)

Sources:
- https://scikit-posthocs.readthedocs.io/en/latest/intro.html

- https://seaborn.pydata.org/generated/seaborn.heatmap.html

- https://seaborn.pydata.org/examples/many_pairwise_correlations.html

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scikit_posthocs import posthoc_dunn
from scipy.stats import kruskal
import numpy as np

# Function to generate a significance heatmap for a single statistic
def plot_pairwise_significance_heatmap(df, stat_to_compare, cluster_col, alpha=0.05):
    """
    Performs Kruskal-Wallis and Dunn's post-hoc test, then plots the
    pairwise p-values as a heatmap.
    """
    print(f"Generating Significance Heatmap for: {stat_to_compare}")

    # Kruskal-Wallis H-test
    cluster_data = [group[stat_to_compare].dropna() for name, group in df.groupby(cluster_col)]

    kruskal_h, kruskal_p = kruskal(*cluster_data)

    if kruskal_p >= alpha:
        print(f"Kruskal-Wallis test is NOT significant (p={kruskal_p:.4f}). No pairwise differences to plot.")
        return

    print(f"Kruskal-Wallis test is significant (p={kruskal_p:.4f}). Generating heatmap...")

    # Dunn's post-hoc test for pairwise comparisons
    dunn_p_values = posthoc_dunn(df, p_adjust='bonferroni', val_col=stat_to_compare, group_col=cluster_col)

    # Create a mask to hide the upper triangle (since the matrix is symmetrical)
    mask = np.triu(np.ones_like(dunn_p_values, dtype=bool))

    # Create the Heatmap
    plt.figure(figsize=(10, 8))
    sns.heatmap(
        dunn_p_values,
        annot=True,              # Show the p-values in the cells
        fmt=".3f",               # Format to 3 decimal places
        cmap='Reds_r',           # Reversed colormap: low p-values (significant) are darker red
        mask=mask,               # Apply the mask to hide the upper triangle
        linewidths=.5,
        vmin=0, vmax=alpha       # Anchor the colormap from 0 to your alpha for emphasis
    )

    plt.title(f"Pairwise Significance (p-values) for '{stat_to_compare}'\n(Bonferroni Correction)", fontsize=14)
    plt.xlabel("Cluster")
    plt.ylabel("Cluster")
    plt.show()

# Define the columns you want to SKIP
columns_to_exclude = ['Player', 'Season', 'Team','LEMS_Occurred', 'injury_cluster'] + feature_columns_for_clustering


# Loop through ALL columns
for column_name in lems_seasons_df_copy.columns:
    # if statement to skip unwanted columns
    if column_name not in columns_to_exclude:
        # call the function for each player stat
        plot_pairwise_significance_heatmap(lems_seasons_df_copy, column_name, 'injury_cluster')
```

```python
# This column is True if any of the relevant flags are True for an injury record
existing_relevant_flags = [col for col in lems_precursor_flag_columns_in_df if col in new_injury_data_cleaned.columns]
if not existing_relevant_flags:
    raise ValueError("None of the lems_precursor_flag_columns_in_df found in new_injury_data_cleaned. Please check column names.")
if len(existing_relevant_flags) < len(lems_precursor_flag_columns_in_df):
    print(f"Warning: Not all lems_precursor_flag_columns_in_df found. Using: {existing_relevant_flags}")

new_injury_data_cleaned['is_count_relevant_prior_injury'] = new_injury_data_cleaned[existing_relevant_flags].any(axis=1)

# Determine Season for LEMS Events in new_df_lems_cases
def get_season(date):
    if pd.isna(date):
        return None
    if date.month >= 8:
        return f"{date.year}-{date.year + 1}"
    else:
        return f"{date.year - 1}-{date.year}"

new_df_lems_cases['Season'] = new_df_lems_cases['Date'].apply(get_season)
# Drop LEMS cases where season could not be determined (if any)
new_df_lems_cases.dropna(subset=['Player', 'Season', 'Date'], inplace=True)


# Find First LEMS Date per Player-Season
first_lems_dates_df = new_df_lems_cases.loc[
    new_df_lems_cases.groupby(['Player', 'Season'])['Date'].idxmin()
][['Player', 'Season', 'Date']].rename(columns={'Date': 'First_LEMS_Date_In_Season'})

# Calculate previous_injury_count
player_season_counts = []

for _, row in first_lems_dates_df.iterrows():
    player = row['Player']
    season = row['Season']
    first_lems_date = row['First_LEMS_Date_In_Season']

    # Filter new_injury_data_cleaned for this player's injuries BEFORE their first LEMS in this season
    prior_injuries = new_injury_data_cleaned[
        (new_injury_data_cleaned['Player'] == player) &
        (new_injury_data_cleaned['Date'] < first_lems_date) &
        (new_injury_data_cleaned['is_count_relevant_prior_injury'] == True) # Count only relevant types
    ]

    count = len(prior_injuries)
    player_season_counts.append({
        'Player': player,
        'Season': season,
        'previous_injury_count': count
    })

prior_counts_df = pd.DataFrame(player_season_counts)

# Merge into stats_df
if not prior_counts_df.empty:
    # Assuming stats_df has 'Player' and 'Season' columns matching the format
    new_stats_df = pd.merge(new_stats_df, prior_counts_df, on=['Player', 'Season'], how='left')
    # Fill NaN with 0 for players/seasons in stats_df that didn't have a LEMS
    # or had no prior relevant injuries before their first LEMS.
    new_stats_df['previous_injury_count'] = new_stats_df['previous_injury_count'].fillna(0).astype(int)
else:
    # If no LEMS events or no prior injuries, add the column with zeros
    new_stats_df['previous_injury_count'] = 0
    print("No prior injury counts were calculated (prior_counts_df is empty). 'previous_injury_count' column added as 0 to new_stats_df.")


print("Finished calculating and merging 'previous_injury_count' into new_stats_df.")
# print(stats_df.head())
# print(stats_df[stats_df['previous_injury_count'] > 0].head())
```

    - The "Superstar" (Cluster 5) and "Veteran / Back Instability" (Cluster 3) profiles had significantly more Assists and Turnovers, validating them as primary playmakers.
    - The "Ankle Precursor" (Cluster 1) and "Quad Recurrence" (Cluster 2) profiles had significantly more Rebounds than the "First-Time LEMS" group, indicating a more active interior presence.
    - The "Concussion-Linked" (Cluster 4) profile had significantly fewer Steals, distinguishing its defensive activity from the veteran and star profiles.
  - Conclusion: How a player plays matters. A high Free Throw Attempt Rate suggests a physical, driving style that might correlate with groin or quad issues. A high Three_Point_Attempt_Rate points to a perimeter game that may stress hamstrings differently. This analysis shows that a player's on-court role and style are intertwined with their specific LEMS pathway.

4. Player Value & Efficiency
  - Finding: Advanced metrics like Player Efficiency Rating (PER), Box Plus-Minus (BPM), and Win Shares robustly separated the clusters. The "Superstar / Hamstring Recurrence" profile (Cluster 5) and the "High-Motor / Ankle Precursor" profile (Cluster 1) both showed significantly higher overall value than the "First-Time LEMS" group (Cluster 0).
  - Conclusion: Player caliber matters. The context of a LEMS is different for a highly efficient, high-value player compared to a role player. This shows that elite performance and specific injury vulnerabilities can coexist and define a unique risk profile.
